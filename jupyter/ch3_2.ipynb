```python
s = requests.Session()
```

```python
s.headers.update({'user-agent' : 'my-crawler/1.0 (+foo@example.com)'})
```

```python
# Sessionオブジェクトには get()、post()などのメソッドがあり、requests.get()、requests.post()などと同様に使える。
r = s.get('https://gihyo.jp/')
r2 = s.get('https://gihyo.jp/dp')
```

```python
r.headers
```
