```python
import lxml.html
```

```python
tree = lxml.html.parse('index.html')
```

```python
# パースすると_ElementTreeオブジェクトが得られる
type(tree)
```

```python
html = tree.getroot()
```

```python
type(html)
```

```python
html.xpath('//li')
```

```python
html.cssselect('li')
```

```python
h1 = html.xpath('//h1')[0]
h1.tag #tag属性でタグの名前を取得できる
```

```python
h1.get('id')
```

```python
h1.getparent()
```

```python
for a in html.cssselect('a'):
    print(a.get('href'), a.text)
```

# Beautiful Soupによるスクレイピング

```python
from bs4 import BeautifulSoup

with open('index.html') as f:
    soup = BeautifulSoup(f, 'html.parser')
    
for a in soup.find_all('a'):
    print(a.get('href'), a.text)
```
